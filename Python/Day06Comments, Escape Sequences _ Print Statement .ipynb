# Escape Sequences


```python
# types of :

# Code	  Result
# \'	  Single Quote
# \\	  Backslash
# \n	  New Line
# \r	  Carriage Return
# \t	  Tab
# \b	  Backspace
# \f	  Form Feed
# \ooo	Octal value
# \xhh	Hex value

print("Type 1 Single Quote")
txt = 'It\'s alright.\n'
print(txt)

print("Type 2 Backslash")
txt = "This will insert one \\ (backslash).\n"
print(txt)

print("Type 3 New line")
txt = "Hello\nWorld!\n"
print(txt)

print("Type 4 Carriage Return")
txt = "Hello\rWorld!\n"
print(txt)

print("Type 5 Tab")
txt = "Hello\tWorld!\n"
print(txt)

print("Type 6 Backspace")
#This example erases one character (backspace):
txt = "Hello \bWorld!\n"
print(txt)
```
