```python
# Built-in Data Types
# In programming, data type is an important concept.

# Variables can store data of different types, and different types can do different things.

# Python has the following data types built-in by default, in these categories:

# Text Type:	    str
# Numeric Types:	int, float, complex
# Sequence Types:	list, tuple, range
# Mapping Type:	  dict
# Set Types:	    set, frozenset
# Boolean Type:	  bool
# Binary Types:	  bytes, bytearray, memoryview
# None Type:	    NoneType
```

```python
# Text Type:	    str
name = 'Aman Munir Mujawar'
print(name)
print("The type of this variable is : ", type(name))
```

```python
# Numeric Types:	int, float, complex

# 1. int

x = 20
#display x:
print(x)
#display the data type of x:
print("The type of this variable is : ",type(x),"\n")

# float

y = 20.5
#display x:
print(y)
#display the data type of y:
print("The type of this variable is : ",type(y),"\n")

# 3.complex4

z = 1j
#display x:
print(z)
#display the data type of x:
print("The type of this variable is : ",type(z),"\n")



```

```python
# Sequence Types:	list, tuple, range

# 1. List

list1 = [[1, 2.5, 3.07],["Apple, Banana, PineApple, Watermelon"], [-4, -0.8]]
print("This is list : ",list1,"\n")

# 2. Tuple
tuble11 = ((1, 2.5, 3.07),("Apple, Banana, PineApple, Watermelon"), (-4, -0.8))
print("This is Tuple : ",tuble11,"\n")

# 3. Range

x = range(6)
#display x:
print(x)
#display the data type of x:
print(type(x))
```
